```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Copy of IT_Service_Requests(1).csv', encoding='ISO-8859-1')
```

```python
df = df.fillna({'FAULTY_EQUIPMENT':'OTHER',
                'PROBLEM_TYPE':'Other - Please specify',
                'NAME_OF_EQUIPMENT':'OTHERS',
                'EQUIPMENT_MAKE':'OTHERS',
                'SERVICE_COMPANY':'OTHERS'})
```

```python
df.dropna(subset=['ID_NO'], inplace=True)
df.dropna(subset=['PROBLEM_DESCRIPTION'], inplace=True)
df.dropna(subset=['REMEDY_DESCRIPTION'], inplace=True)
```

```python
df.info()
```

```python
questions = []
for i in range(0, 50000):
    faulty_equipment = df.iloc[i]['FAULTY_EQUIPMENT']
    problem_type = df.iloc[i]['PROBLEM_TYPE']
    problem_description = df.iloc[i]['PROBLEM_DESCRIPTION']
    name_of_equipment = df.iloc[i]['NAME_OF_EQUIPMENT']
    service_company = df.iloc[i]['SERVICE_COMPANY']

    str_val = "Problem in- " + str(faulty_equipment) + ", Problem type is- " + str(problem_type) + ", Problem description is- " + str(problem_description) + ", Name of Equipment- " + str(name_of_equipment) + ", Service Company is- " + str(service_company)
    questions.append(str_val)
```

```python
answers = []
for i in range(0, 50000):
    remedy_description = df.iloc[i]['REMEDY_DESCRIPTION']
    answers.append(remedy_description)
```
